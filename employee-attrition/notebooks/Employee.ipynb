```python
from sklearn.model_selection import train_test_split

def data_split(data, target, train_frac, random_state=42):
    ''' 
    param data:       Data to be split
    param train_frac: Ratio of train set the datasetfraction

    Randomly split dataset, based on these ratios:
        'train': train_frac
        'valid': (1-train_frac) / 2
        'test':  (1-train_frac) / 2

    Eg: passing train_frac=0.8 gives a 80% / 10% / 10% split
    '''

    assert train_frac >= 0 and train_frac <= 1, "Invalid training set fraction"

    X_train, X_tmp, Y_train, Y_tmp = train_test_split(
                                        data, target, train_size=train_frac, stratify=target,random_state=random_state)

    X_val, X_test, Y_val, Y_test   = train_test_split(
                                        X_tmp, Y_tmp, train_size=0.5, stratify=Y_tmp, random_state=random_state)

    return X_train, X_val, X_test,  Y_train, Y_val, Y_test

```

```python
X_train, X_val, X_test,  Y_train, Y_val, Y_test = data_split(df_features, df_target, train_frac=0.8, random_state=42)
```
